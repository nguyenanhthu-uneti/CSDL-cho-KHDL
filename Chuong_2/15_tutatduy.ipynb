```python
# Import thư viện cần thiết
import sqlite3
# Tạo kết nối đến database
conn = sqlite3.connect('student_datas.db')
cursor = conn.cursor()

# Tạo bảng student_data
cursor.execute('''
    CREATE TABLE IF NOT EXISTS student_data (
        student_id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        major TEXT,
        course_id INTEGER,
        grade REAL,
        FOREIGN KEY (course_id) REFERENCES course_data(course_id)
    )
''')

# Tạo bảng course_data
cursor.execute('''
    CREATE TABLE IF NOT EXISTS course_data (
        course_id INTEGER PRIMARY KEY,
        course_name TEXT NOT NULL
    )
''')

# Dữ liệu mẫu
student_data = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]

course_data = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin Hoc')
]

# Chèn dữ liệu vào bảng course_data
cursor.executemany('INSERT OR REPLACE INTO course_data VALUES (?, ?)', course_data)

# Chèn dữ liệu vào bảng student_data
cursor.executemany('INSERT OR REPLACE INTO student_data VALUES (?, ?, ?, ?, ?)', student_data)

# Lưu thay đổi
conn.commit()

# Kiểm tra dữ liệu trong bảng student_data
print("Student Data:")
cursor.execute('SELECT * FROM student_data')
for row in cursor.fetchall():
    print(row)

# Kiểm tra dữ liệu trong bảng course_data
print("\nCourse Data:")
cursor.execute('SELECT * FROM course_data')
for row in cursor.fetchall():
    print(row)

# Đóng kết nối
conn.close()
```

```python
# Kết nối đến datas
conn = sqlite3.connect('student_datas.db')
cursor = conn.cursor()

# 1. Tích Descartes (Cartesian Product)
print("1. Tích Descartes (Cartesian Product):")
cursor.execute('''
    SELECT s.student_id, s.name, s.major, s.grade, c.course_id, c.course_name
    FROM student_data s, course_data c
''')
for row in cursor.fetchall():
    print(row)
print("\n")
```

```python
# 2. INNER JOIN
print("2. INNER JOIN:")
cursor.execute('''
    SELECT s.student_id, s.name, s.major, s.grade, c.course_id, c.course_name
    FROM student_data s
    INNER JOIN course_data c ON s.course_id = c.course_id
''')
for row in cursor.fetchall():
    print(row)
print("\n")
```

```python
# 3. LEFT JOIN
print("3. LEFT JOIN:")
cursor.execute('''
    SELECT s.student_id, s.name, s.major, s.grade, c.course_id, c.course_name
    FROM student_data s
    LEFT JOIN course_data c ON s.course_id = c.course_id
''')
for row in cursor.fetchall():
    print(row)
print("\n")
```

```python
# 4. RIGHT JOIN
print("4. RIGHT JOIN:")
cursor.execute('''
    SELECT s.student_id, s.name, s.major, s.grade, c.course_id, c.course_name
    FROM student_data s
    RIGHT JOIN course_data c ON s.course_id = c.course_id
''')
for row in cursor.fetchall():
    print(row)
print("\n")
```

```python
print("5. FULL OUTER JOIN (mô phỏng):")
cursor.execute('''
    SELECT s.student_id, s.name, s.major, s.grade, c.course_id, c.course_name
    FROM student_data s
    LEFT JOIN course_data c ON s.course_id = c.course_id
    UNION
    SELECT s.student_id, s.name, s.major, s.grade, c.course_id, c.course_name
    FROM course_data c
    LEFT JOIN student_data s ON s.course_id = c.course_id
    WHERE s.student_id IS NULL
''')
for row in cursor.fetchall():
    print(row)
# Đóng kết nối
conn.close()
```

```python
# Kết nối đến database
conn = sqlite3.connect('student_datas.db')
cursor = conn.cursor()

# 1. Cập nhật và làm sạch dữ liệu
# Cập nhật các giá trị NULL trong course_id bằng cách gán ngẫu nhiên từ course_data
cursor.execute('''
    UPDATE student_data
    SET course_id = (SELECT course_id FROM course_data ORDER BY RANDOM() LIMIT 1)
    WHERE course_id IS NULL
''')

# Commit thay đổi
conn.commit()
```

```python
# a. Tổng số sinh viên, điểm trung bình của từng lớp (major)
print("a. Tổng số sinh viên và điểm trung bình theo lớp:")
cursor.execute('''
    SELECT major, 
           COUNT(*) as student_count, 
           ROUND(AVG(grade), 2) as avg_grade
    FROM student_data
    GROUP BY major
''')
for row in cursor.fetchall():
    print(f"Lớp: {row[0]}, Số SV: {row[1]}, Điểm TB: {row[2]}")
print("\n")
```

```python
# b. Tổng số sinh viên, điểm trung bình của từng môn học
print("b. Tổng số sinh viên và điểm trung bình theo môn học:")
cursor.execute('''
    SELECT c.course_name, 
           COUNT(s.student_id) as student_count, 
           ROUND(AVG(s.grade), 2) as avg_grade
    FROM student_data s
    JOIN course_data c ON s.course_id = c.course_id
    GROUP BY c.course_name
''')
for row in cursor.fetchall():
    print(f"Môn: {row[0]}, Số SV: {row[1]}, Điểm TB: {row[2]}")
print("\n")
```

```python
# c. Phân loại thi đua theo điểm trung bình của từng môn học
print("c. Phân loại thi đua theo môn học:")
cursor.execute('''
    SELECT c.course_name,
           ROUND(AVG(s.grade), 2) as avg_grade,
           CASE 
               WHEN AVG(s.grade) >= 9.0 THEN 'Xuất sắc'
               WHEN AVG(s.grade) >= 6.0 THEN 'Tốt'
               ELSE 'Kém'
           END as classification
    FROM student_data s
    JOIN course_data c ON s.course_id = c.course_id
    GROUP BY c.course_name
''')
for row in cursor.fetchall():
    print(f"Môn: {row[0]}, Điểm TB: {row[1]}, Xếp loại: {row[2]}")

# Đóng kết nối
conn.close()
```

```python
# Kết nối đến database
conn = sqlite3.connect('student_datas.db')
cursor = conn.cursor()

# a. Xếp hạng sinh viên theo điểm số (toàn bộ)
print("a. Xếp hạng sinh viên theo điểm số (toàn bộ):")
cursor.execute('''
    SELECT 
        student_id,
        name,
        major,
        grade,
        RANK() OVER (ORDER BY grade DESC) as rank,
        DENSE_RANK() OVER (ORDER BY grade DESC) as dense_rank
    FROM student_data
    ORDER BY grade DESC
''')
print("Student_ID | Name                 | Major      | Grade | Rank | Dense_Rank")
for row in cursor.fetchall():
    print(f"{row[0]:<10} | {row[1]:<20} | {row[2]:<10} | {row[3]:<5} | {row[4]:<4} | {row[5]}")
print("\n")
```

```python
# b. Xếp hạng sinh viên theo điểm số trong từng lớp (major)
conn = sqlite3.connect('student_datas.db')
cursor = conn.cursor()
print("b. Xếp hạng sinh viên theo điểm số trong từng lớp:")
cursor.execute('''
    SELECT 
        student_id,
        name,
        major,
        grade,
        RANK() OVER (PARTITION BY major ORDER BY grade DESC) as rank_in_major
    FROM student_data
    ORDER BY major, grade DESC
''')
print("Student_ID | Name                 | Major      | Grade | Rank ")
for row in cursor.fetchall():
    print(f"{row[0]:<10} | {row[1]:<20} | {row[2]:<10} | {row[3]:<5} | {row[4]}")
print("\n")
conn.close()
```

```python
# c. Xếp hạng sinh viên theo điểm số trong từng môn học
conn = sqlite3.connect('student_datas.db')
cursor = conn.cursor()
print("c. Xếp hạng sinh viên theo điểm số trong từng môn học:")
cursor.execute('''
    SELECT 
        s.student_id,
        s.name,
        c.course_name,
        s.grade,
        RANK() OVER (PARTITION BY s.course_id ORDER BY s.grade DESC) as rank_in_course
    FROM student_data s
    JOIN course_data c ON s.course_id = c.course_id
    ORDER BY c.course_name, s.grade DESC
''')
print("Student_ID | Name                 | Major           | Grade | Rank ")
for row in cursor.fetchall():
    print(f"{row[0]:<10} | {row[1]:<20} | {row[2]:<15} | {row[3]:<5} | {row[4]}")
# Đóng kết nối
conn.close()
```

```python
# Kết nối đến database
conn = sqlite3.connect('student_datas.db')
cursor = conn.cursor()

# a. Xếp hạng theo điểm số (toàn bộ sinh viên)
print("a. Xếp hạng theo điểm số (toàn bộ):")
cursor.execute('''
    SELECT student_id, name, major, grade,
           RANK() OVER (ORDER BY grade DESC) as rank
    FROM student_data
''')

print("Top 3 cao nhất:")
cursor.execute('''
    SELECT student_id, name, major, grade
    FROM student_data
    ORDER BY grade DESC
    LIMIT 3
''')
for row in cursor.fetchall():
    print(f"ID: {row[0]}, Tên: {row[1]}, Lớp: {row[2]}, Điểm: {row[3]}")

print("\nTop 3 thấp nhất:")
cursor.execute('''
    SELECT student_id, name, major, grade
    FROM student_data
    ORDER BY grade ASC
    LIMIT 3
''')
for row in cursor.fetchall():
    print(f"ID: {row[0]}, Tên: {row[1]}, Lớp: {row[2]}, Điểm: {row[3]}")
print("\n")
```

```python
# b. Xếp hạng theo điểm số theo lớp học (major)
print("b. Xếp hạng theo điểm số theo lớp học:")
cursor.execute('''
    SELECT student_id, name, major, grade,
           RANK() OVER (PARTITION BY major ORDER BY grade DESC) as rank
    FROM student_data
''')

# Lấy top 3 cao nhất và thấp nhất cho từng lớp
majors = ['May Tinh', 'Kinh Te', 'Toan Tin']
for major in majors:
    print(f"Lớp: {major}")
    
    print("Top 3 cao nhất:")
    cursor.execute('''
        SELECT student_id, name, grade
        FROM student_data
        WHERE major = ?
        ORDER BY grade DESC
        LIMIT 3
    ''', (major,))
    for row in cursor.fetchall():
        print(f"ID: {row[0]}, Tên: {row[1]}, Điểm: {row[2]}")
        
    print("Top 3 thấp nhất:")
    cursor.execute('''
        SELECT student_id, name, grade
        FROM student_data
        WHERE major = ?
        ORDER BY grade ASC
        LIMIT 3
    ''', (major,))
    for row in cursor.fetchall():
        print(f"ID: {row[0]}, Tên: {row[1]}, Điểm: {row[2]}")
    print("\n")
```

```python
# c. Xếp hạng theo điểm số theo mã môn học
print("c. Xếp hạng theo điểm số theo mã môn học:")
cursor.execute('''
    SELECT s.student_id, s.name, s.grade, c.course_id, c.course_name,
           RANK() OVER (PARTITION BY c.course_id ORDER BY s.grade DESC) as rank
    FROM student_data s
    JOIN course_data c ON s.course_id = c.course_id
''')

# Lấy top 3 cao nhất và thấp nhất cho từng môn
cursor.execute('SELECT course_id, course_name FROM course_data')
courses = cursor.fetchall()

for course in courses:
    course_id, course_name = course
    print(f"Môn: {course_name} (ID: {course_id})")
    
    print("Top 3 cao nhất:")
    cursor.execute('''
        SELECT s.student_id, s.name, s.grade
        FROM student_data s
        WHERE s.course_id = ?
        ORDER BY s.grade DESC
        LIMIT 3
    ''', (course_id,))
    for row in cursor.fetchall():
        print(f"ID: {row[0]}, Tên: {row[1]}, Điểm: {row[2]}")
        
    print("Top 3 thấp nhất:")
    cursor.execute('''
        SELECT s.student_id, s.name, s.grade
        FROM student_data s
        WHERE s.course_id = ?
        ORDER BY s.grade ASC
        LIMIT 3
    ''', (course_id,))
    for row in cursor.fetchall():
        print(f"ID: {row[0]}, Tên: {row[1]}, Điểm: {row[2]}")
    print("\n")

# Đóng kết nối
conn.close()
```
