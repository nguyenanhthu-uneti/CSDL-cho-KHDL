# 1. Kết nối 2 bảng trên bằng cách sử dụng tích Descartes :

```python
#Kết nối lại SQLite 
conn = sqlite3.connect("students.db")

#Thực hiện tích Decartes giữa student và course :
query = """
SELECT *
FROM student, course ;
"""
df_result = pd.read_sql_query(query, conn)

#Hiển thị kết quả 
print("Kết quả tích Descartes giữa student và course:")
display(df_result)
```

# Kết nối 2 bảng trên bằng cách sử dụng JOIN : INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL OUTER JOIN

```python
#INNER JOIN 
query_inner = """
SELECT student.*, course.course_name
FROM student
INNER JOIN course ON student.course_id = course.id ;
"""
df_inner = pd.read_sql_query(query_inner, conn)
print("INNER JOIN:")
display(df_inner)

#LEFT JOIN
query_left = """
SELECT student.*, course.course_name
FROM student
LEFT JOIN course ON student.course_id = course.id ;
"""
df_left = pd.read_sql_query(query_left, conn)
print("LEFT JOIN:")
display(df_left)

#RIGHT JOIN (ko được hỗ trợ trực tiếp trong sqlite nên e đổi vị trí bảng rồi dùng LEFT JOIN ạ)
query_right = """
SELECT course.*, student.name, student.class, student.score
FROM course
LEFT JOIN student ON student.course_id = course.id ;
"""
df_right = pd.read_sql_query(query_right, conn)
print("RIGHT JOIN (mô phỏng bằng LEFT JOIN):")
display(df_right)

#FULL OUTER JOIN (ko được hỗ trợ trực tiếp, làm bằng UNION của LEFT JOIN và RIGHT JOIN)
query_full_outer = """
SELECT student.*, course.course_name
FROM student
LEFT JOIN course ON student.course_id = course.id
UNION
SELECT student.*, course.course_name
FROM course
LEFT JOIN student ON student.course_id = course.id;
"""
df_full_outer = pd.read_sql_query(query_full_outer, conn)
print("FULL OUTER JOIN (dùng UNION):")
display(df_full_outer)
```

# 2. Cập nhật những giá trị course_id còn thiếu trong bảng student bằng câu lệnh SQL, trong đó các giá trị được điền là những giá trị nằm trong bảng course và loại bỏ những bản ghi tham gia những môn học không tồn tại bảng course

```python
# Kết nối lại với database
conn = sqlite3.connect("students.db")  
cursor = conn.cursor()

# Lấy danh sách course_id có trong bảng course
cursor.execute("SELECT id FROM course")
valid_courses = [row[0] for row in cursor.fetchall()]

# Cập nhật course_id còn thiếu bằng giá trị ngẫu nhiên từ course có sẵn
if valid_courses:
    # Lấy danh sách các sinh viên có course_id NULL
    cursor.execute("SELECT student_id FROM student WHERE course_id IS NULL")
    students_with_null = [row[0] for row in cursor.fetchall()]
    
    # Gán giá trị ngẫu nhiên cho từng sinh viên bị thiếu course_id
    import random
    for student_id in students_with_null:
        random_course = random.choice(valid_courses)
        cursor.execute("UPDATE student SET course_id = ? WHERE student_id = ?", (random_course, student_id))

    conn.commit()

# Xóa các sinh viên có course_id không tồn tại trong bảng course
cursor.execute("""
    DELETE FROM student 
    WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()

# Đọc dữ liệu sau khi cập nhật
df_student = pd.read_sql_query("SELECT * FROM student;", conn)

# Hiển thị kết quả
print("Dữ liệu bảng student sau khi cập nhật:")
display(df_student)
```

# a. Tổng số sinh viên, điểm trung bình của từng lớp 

```python
query_class_stats = """
SELECT class, COUNT(*) AS total_students, AVG(score) AS avg_score
FROM student
GROUP BY class;
"""

df_class_stats = pd.read_sql_query(query_class_stats, conn)
print("Tổng số sinh viên và điểm trung bình theo lớp:")
display(df_class_stats)
```

# b. Tổng số sinh viên, điểm trung bình của từng môn học

```python
query_course_stats = """
SELECT course.course_name, COUNT(student.student_id) AS total_students, AVG(student.score) AS avg_score
FROM student
JOIN course ON student.course_id = course.id
GROUP BY course.course_name;
"""

df_course_stats = pd.read_sql_query(query_course_stats, conn)
print(" Tổng số sinh viên và điểm trung bình theo môn học:")
display(df_course_stats)
```

# c. Phân loại thi đua theo số điểm của từng môn học biết :
* i. Điểm TB >= 9.0 : Xuất sắc.
* ii. 6.0 <= Điểm TB <= 8.9 : Tốt.
* iii. Điểm TB < 6.0 : Kém.

```python
query_ranking = """
SELECT course.course_name, 
       COUNT(student.student_id) AS total_students, 
       AVG(student.score) AS avg_score,
       CASE 
           WHEN AVG(student.score) >= 9.0 THEN 'Xuất sắc'
           WHEN AVG(student.score) >= 6.0 THEN 'Tốt'
           ELSE 'Kém'
       END AS ranking
FROM student
JOIN course ON student.course_id = course.id
GROUP BY course.course_name;
"""

df_ranking = pd.read_sql_query(query_ranking, conn)
print("Phân loại thi đua theo từng môn học:")
display(df_ranking)
```

# 3. Xếp hạng sinh viên theo :

a, Điểm số

```python
# Kết nối SQLite
conn = sqlite3.connect("students.db") 

# Xếp hạng theo điểm số toàn bộ sinh viên với DENSE_RANK()
query_rank_score = """
SELECT student_id, name, class, course_id, score,
       DENSE_RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student;
"""
df_rank_score = pd.read_sql_query(query_rank_score, conn)

print("Xếp hạng sinh viên theo điểm số:")
display(df_rank_score)
```

b, Điểm số theo lớp học

```python
# Xếp hạng theo điểm số trong từng lớp học bằng DENSE_RANK()
query_dense_rank_class = """
SELECT student_id, name, class, course_id, score,
       DENSE_RANK() OVER (PARTITION BY class ORDER BY score DESC) AS dense_rank_class
FROM student;
"""
df_dense_rank_class = pd.read_sql_query(query_dense_rank_class, conn)
print("Xếp hạng sinh viên theo điểm số trong từng lớp (DENSE_RANK()):")
display(df_dense_rank_class)
```

c, Điểm số theo mã môn học

```python
# Xếp hạng theo điểm số trong từng môn học
query_rank_course = """
SELECT student_id, name, class, course_id, score,
       DENSE_RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_course
FROM student;
"""
df_rank_course = pd.read_sql_query(query_rank_course, conn)

print("Xếp hạng sinh viên theo điểm số trong từng môn học:")
display(df_rank_course)
```

```python
print("Top 3 sinh viên cao nhất theo điểm số:")
display(df_rank_score.nlargest(3, 'score'))  # Chọn sinh viên có điểm số cao nhất

print("Top 3 sinh viên thấp nhất theo điểm số:")
display(df_rank_score.nsmallest(3, 'score'))  # Chọn sinh viên có điểm số thấp nhất
```

```python
# Kết nối SQLite
conn = sqlite3.connect("students.db") 
# Top 3 sinh viên có điểm cao nhất/thấp nhất theo từng lớp
def get_top_n_students_by_class(n=3, order='DESC'):
    query = f'''
    WITH RankedStudents AS (
        SELECT student_id, name, class, course_id, score,
               DENSE_RANK() OVER (PARTITION BY class ORDER BY score {'DESC' if order == 'DESC' else 'ASC'}) AS rank
        FROM student
    )
    SELECT student_id, name, class, course_id, score
    FROM RankedStudents
    WHERE rank <= {n}
    ORDER BY class, score {order};
    '''
    return pd.read_sql_query(query, conn)

# Lấy top 3 sinh viên có điểm cao nhất/thấp nhất theo từng lớp
top_3_by_class_highest = get_top_n_students_by_class(3, 'DESC')
top_3_by_class_lowest = get_top_n_students_by_class(3, 'ASC')

print("\nTop 3 sinh viên có điểm cao nhất theo từng lớp:")
print(top_3_by_class_highest)

print("\nTop 3 sinh viên có điểm thấp nhất theo từng lớp:")
print(top_3_by_class_lowest)
```

```python
# Top 3 sinh viên cao nhất theo từng môn học
print("Top 3 sinh viên cao nhất theo từng môn học:")
display(df_rank_course.groupby('course_id', group_keys=False)
       .apply(lambda x: x.nsmallest(3, 'rank_course'), include_groups=False)
       .reset_index(drop=True))

# Top 3 sinh viên thấp nhất theo từng môn học
print("Top 3 sinh viên thấp nhất theo từng môn học:")
display(df_rank_course.groupby('course_id', group_keys=False)
       .apply(lambda x: x.nlargest(3, 'rank_course'), include_groups=False)
       .reset_index(drop=True))

# Đóng kết nối
conn.close()
```

```python
from datetime import datetime, timedelta

# Kết nối database
conn = sqlite3.connect("students.db")
cursor = conn.cursor()

# Thêm cột graduation_date nếu chưa có
try:
    cursor.execute("ALTER TABLE Student ADD COLUMN graduation_date DATETIME")
except sqlite3.OperationalError:
    print("Cột graduation_date đã tồn tại, tiếp tục cập nhật dữ liệu.")

# Lấy dữ liệu sinh viên
df = pd.read_sql("SELECT student_id, score FROM Student", conn)

# Xếp hạng theo score (điểm cao hơn -> rank thấp hơn)
df["rank"] = df["score"].rank(method="dense", ascending=False)

# Cập nhật graduation_date dựa trên rank
for index, row in df.iterrows():
    graduation_date = datetime.now() + timedelta(days=row["rank"] * 180)  # Thêm 6 tháng cho mỗi bậc
    graduation_date_str = graduation_date.strftime("%Y-%m-%d %H:%M:%S")  # Chuyển datetime -> chuỗi
    cursor.execute("UPDATE Student SET graduation_date = ? WHERE student_id = ?", (graduation_date_str, row["student_id"]))

# Lưu thay đổi 
conn.commit()

# Đọc toàn bộ bảng Student sau khi cập nhật
df_after_update = pd.read_sql("SELECT * FROM Student", conn)

# Hiển thị kết quả
print(df_after_update)

# Đóng kết nối
conn.close()
```
