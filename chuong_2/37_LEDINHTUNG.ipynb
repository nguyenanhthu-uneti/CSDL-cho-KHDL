## 1. Hãy kết nối hai bảng trên theo những cách sau:
- Sử dụng tích descartes
- Sử dụng JOIN

```python
# Tích Descartes (CROSS JOIN)
query = "SELECT * FROM student CROSS JOIN course;"
df_cross_join = pd.read_sql_query(query, conn)
df_cross_join
```

```python
# INNER JOIN
query = """
SELECT student.student_id, student.name, student.class, student.course_id, student.score, course.course_name
FROM student
INNER JOIN course ON student.course_id = course.id;
"""
df_inner_join = pd.read_sql_query(query, conn)
df_inner_join
```

```python
# LEFT JOIN
query = """
SELECT student.student_id, student.name, student.class, student.course_id, student.score, course.course_name
FROM student
LEFT JOIN course ON student.course_id = course.id;
"""
df_left_join = pd.read_sql_query(query, conn)
df_left_join
```

```python
# RIGHT JOIN
query = """
SELECT course.id, course.course_name, student.student_id, student.name, student.class, student.score
FROM course
LEFT JOIN student ON student.course_id = course.id;
"""
df_right_join = pd.read_sql_query(query, conn)
df_right_join
```

```python
query = """
SELECT student.student_id, student.name, student.class, student.course_id, student.score, course.course_name
FROM student
LEFT JOIN course ON student.course_id = course.id

UNION

SELECT student.student_id, student.name, student.class, student.course_id, student.score, course.course_name
FROM student
RIGHT JOIN course ON student.course_id = course.id;
"""
df_full_outer_join = pd.read_sql_query(query, conn)
df_full_outer_join
```

## 2. Hãy cập nhật những giá trị ***course_id*** còn thiếu trong bảng ***student*** bằng câu lệnh SQL, trong đó các giá trị được điền là những giá trị nằm trong bảng ***course*** và loại bỏ những bản ghi tham gia những môn học không tồn tại bảng ***course***. Sau đó hãy cho biết:
1. Tổng số sinh viên, điểm trung bình của từng lớp
2. Tổng số sinh viên, điểm trung bình của từng môn học
3. Phân loại thi đua theo số điểm của từng môn học biết:
- Điểm $TB \geq 9.0$: Xuất sắc
- $8.9 \geq Điểm TB \geq 6.0$: Tốt
- $Điểm TB < 6$: Kém 

```python
# Cập nhật course_id còn thiếu
cursor.execute("""
    UPDATE student
    SET course_id = (SELECT id FROM course WHERE course_name = 'Tin hoc')
    WHERE course_id IS NULL;
""")

# Xóa các sinh viên có course_id không hợp lệ
cursor.execute("""
    DELETE FROM student
    WHERE course_id NOT IN (SELECT id FROM course);
""")
```

```python
### 1. Tính tổng số sinh viên, điểm trung bình của từng lớp
query_class = """
SELECT class, COUNT(student_id) AS total_students, AVG(score) AS avg_score
FROM student
GROUP BY class;
"""
df_class = pd.read_sql_query(query_class, conn)
df_class
```

```python
### 2. Tính tổng số sinh viên, điểm trung bình của từng môn học
query_course = """
SELECT course.course_name, COUNT(student.student_id) AS total_students, AVG(student.score) AS avg_score
FROM student
JOIN course ON student.course_id = course.id
GROUP BY course.course_name;
"""
df_course = pd.read_sql_query(query_course, conn)
df_course
```

```python
### 3. Phân loại sinh viên theo điểm trung bình
query_avg = '''
    SELECT Course.course_name, 
           AVG(Student.score) AS avg_score,
           CASE 
               WHEN AVG(Student.score) >= 9.0 THEN 'Xuất sắc'
               WHEN AVG(Student.score) BETWEEN 6.0 AND 8.9 THEN 'Tốt'
               ELSE 'Kém'
           END AS classification
    FROM Student
    JOIN Course ON Student.course_id = Course.id
    GROUP BY Course.course_name
'''
df_avg = pd.read_sql_query(query_avg, conn)
print("Phân loại thi đua theo số điểm của từng môn học:")
df_avg
```

## 3. Hãy xếp hạng sinh viên thông qua:
- a. Điểm số
- b. Điểm số theo lớp học
- c. Điểm số theo mã môn học

Và cho biết top 3 sinh viên đạt thứ hạng cao nhất, top 3 sinh viên đạt thứ hạng thấp nhất theo từng trường hợp trên

```python
### 1. Xếp hạng sinh viên theo điểm số chung
query = """
SELECT student_id, name, class, course_id, score,
       RANK() OVER (ORDER BY score DESC) AS rank_overall
FROM student;
"""
df_rank_overall = pd.read_sql_query(query, conn)
print('Xếp hạng sinh viên theo điểm số chung:\n',df_rank_overall)

query_top_high = '''
    SELECT student_id, name, class, score
    FROM Student
    ORDER BY score DESC
    LIMIT 3
'''
df_top_high = pd.read_sql_query(query_top_high, conn)
print("Top 3 sinh viên có điểm cao nhất:")
print(df_top_high)

query_top_lowerdf_top_lower = '''
    SELECT student_id, name, class, score
    FROM Student
    ORDER BY score ASC
    LIMIT 3
'''
df_top_lower = pd.read_sql_query(query_top_lowerdf_top_lower, conn)
print("Top 3 sinh viên có điểm thấp nhất:")
print(df_top_lower)
```

```python
### 2. Xếp hạng sinh viên theo điểm số trong từng lớp học
query_class_high = '''
    SELECT student_id, name, class, score
    FROM (
        SELECT student_id, name, class, score,
               RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_high
        FROM Student
    ) WHERE rank_high <= 3
'''
df_class_high = pd.read_sql_query(query_class_high, conn)
print("Top 3 sinh viên đạt thứ hạng cao nhất theo từng lớp:")
print(df_class_high)

query_class_lower = '''
    SELECT student_id, name, class, score
    FROM (
        SELECT student_id, name, class, score,
               RANK() OVER (PARTITION BY class ORDER BY score ASC) AS rank_low
        FROM Student
    ) WHERE rank_low <= 3
'''
df_class_lower = pd.read_sql_query(query_class_lower, conn)
print("Top 3 sinh viên đạt thứ hạng thấp nhất theo từng lớp:")
print(df_class_lower)
```

```python
### 3. sinh viên đạt thứ hạng cao nhất theo từng môn học
query_course_high = '''
    SELECT student_id, name, course_id, score
    FROM (
        SELECT student_id, name, course_id, score,
               RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_high
        FROM Student
    ) WHERE rank_high <= 3
'''
df_course_high = pd.read_sql_query(query_course_high, conn)
print("Top 3 sinh viên đạt thứ hạng cao nhất theo từng môn học:")
print(df_course_high)

query_course_lower = '''
    SELECT student_id, name, course_id, score
    FROM (
        SELECT student_id, name, course_id, score,
               RANK() OVER (PARTITION BY course_id ORDER BY score ASC) AS rank_low
        FROM Student
    ) WHERE rank_low <= 3
'''
df_course_lower = pd.read_sql_query(query_course_lower, conn)
print("Top 3 sinh viên đạt thứ hạng thấp nhất theo từng môn học:")
print(df_course_lower)
```

## 4. Hãy bổ sung thêm trường ***graduation_date*** có kiểu dữ liệu DATETIME vào bảng ***student*** để xác định thời gian tốt nghiệp của từng bạn, trong đó thời gian tốt nghiệp dược xác định bởi thời gian hiện tại cộng với số hạng tương ứng của bạn đó tính theo điểm số

```python
# Thêm cột graduation_date
cursor.execute("""
    ALTER TABLE student
    ADD COLUMN graduation_date DATETIME;
""")
conn.commit()

# Tạo bảng tạm để xếp hạng sinh viên theo điểm số chung
query_rank = """
    CREATE TEMP TABLE student_rank AS
    SELECT student_id,
           RANK() OVER (ORDER BY score DESC) AS rank
    FROM student;
"""
cursor.execute(query_rank)
conn.commit()

# Cập nhật graduation_date dựa trên thứ hạng
cursor.execute("""
    UPDATE student
    SET graduation_date = DATE('now', '+' || (
        SELECT rank
        FROM student_rank
        WHERE student_rank.student_id = student.student_id
    ) || ' days')
""")
conn.commit()

# Kiểm tra bảng student sau khi cập nhật graduation_date
df_student_grad = pd.read_sql_query("SELECT * FROM student;", conn)
print("Bảng student sau khi thêm và cập nhật graduation_date:")
df_student_grad
```
