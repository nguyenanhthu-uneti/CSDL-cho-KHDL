```python
import sqlite3
import pandas as pd

# Kết nối SQLite
conn = sqlite3.connect("student_course.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS student (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score FLOAT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS course (
    id INTEGER PRIMARY KEY,
    course_name TEXT
)
""")

conn.commit()
```

# 1.Kết nối hai bảng


```python
# Tích descartes
df = pd.read_sql_query("""
SELECT * FROM student CROSS JOIN course
""", conn)
df
```

## Sử dụng Join

```python
# INNER JOIN
df = pd.read_sql_query("""
SELECT student.*, course.course_name
FROM student
INNER JOIN course ON student.course_id = course.id
""", conn)
df
```

    Chỉ lấy những bản ghi có course_id trong bảng student khớp với id của course.

```python
# LEFT JOIN
df = pd.read_sql_query("""
SELECT student.*, course.course_name
FROM student
LEFT JOIN course ON student.course_id = course.id
""", conn)
df
```

    Lấy tất cả sinh viên, dù có hoặc không có môn học.

```python
# RIGHT JOIN
df = pd.read_sql_query("""
SELECT student.*, course.course_name
FROM course
LEFT JOIN student ON student.course_id = course.id
""", conn)
df
```

# 2.Cập nhật dữ liệu và loại bỏ các bản ghi không hợp lệ


```python
# Cập nhật course_id bị thiếu bằng giá trị hợp lệ từ bảng course
cursor.execute("""
UPDATE student
SET course_id = (SELECT id FROM course ORDER BY RANDOM() LIMIT 1)
WHERE course_id IS NULL
""")
conn.commit()
```

```python
# Xóa các sinh viên có course_id không tồn tại trong bảng course
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()
```

## Tổng số sinh viên, điểm trung bình của từng lớp

```python
df = pd.read_sql_query("""
SELECT class, COUNT(*) AS total_students, AVG(score) AS avg_score
FROM student
GROUP BY class
""",conn)
df
```

## Tổng số sinh viên, điểm trung bình của từng môn học

```python
df = pd.read_sql_query("""
SELECT course.course_name, COUNT(*) AS total_students, AVG(student.score) AS avg_score
FROM student
JOIN course ON student.course_id = course.id
GROUP BY course.course_name
""",conn)
df
```

## Phân loại sinh viên theo điểm số

```python
df = pd.read_sql_query("""
SELECT course.course_name,
       AVG(student.score) AS avg_score,
       CASE 
           WHEN AVG(student.score) >= 9.0 THEN 'Xuất sắc'
           WHEN AVG(student.score) >= 6.0 THEN 'Tốt'
           ELSE 'Kém'
       END AS rating
FROM student
JOIN course ON student.course_id = course.id
GROUP BY course.course_name
""", conn)
df
```

# 3. Xếp hạng sinh viên

```python
# Xếp hạng theo điểm số
df = pd.read_sql_query("""
SELECT name, score, RANK() OVER (ORDER BY score DESC) AS rank
FROM student
""",conn)
top_3_highest = df.nsmallest(3, 'rank')
top_3_lowest = df.nlargest(3, 'rank')

print("top 3 cao nhất:\n",top_3_highest ,'\n')
print("top 3 thấp nhất:\n",top_3_lowest)
```

```python
# Xếp hạng theo lớp 
df = pd.read_sql_query("""
SELECT class, name, score, RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank
FROM student
""",conn)
top_3_highest = df.nsmallest(3, 'rank')
top_3_lowest = df.nlargest(3, 'rank')

print("top 3 cao nhất:\n",top_3_highest ,'\n')
print("top 3 thấp nhất:\n",top_3_lowest)
```

```python
# Xếp hạng theo môn học
df = pd.read_sql_query("""
SELECT course.course_name, name, score, RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank
FROM student
JOIN course ON student.course_id = course.id
""", conn)
top_3_highest = df.nsmallest(3, 'rank')
top_3_lowest = df.nlargest(3, 'rank')

print("top 3 cao nhất:\n",top_3_highest ,'\n')
print("top 3 thấp nhất:\n",top_3_lowest)
```
