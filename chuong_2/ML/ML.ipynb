```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Bước 1: Tạo bộ dữ liệu giả lập
np.random.seed(42)
n_samples = 200
beta_0, beta_1, beta_2 = 7, 3.5, -2.1

X1 = np.random.uniform(-5, 5, n_samples)
X2 = np.random.uniform(-5, 5, n_samples)
epsilon = np.random.normal(0, 2, n_samples)

Y = beta_0 + beta_1 * X1 + beta_2 * X2 + epsilon
print(Y)
```

```python
# Định nghĩa mô hình
class Model:
    def __init__(self):
        self.W = np.random.randn(3)  # (w1, w2, b)
        self.v = np.zeros(3)  # Momentum
        self.s = np.zeros(3)  # RMSprop
        self.m = np.zeros(3)  # Adam
        self.beta1 = 0.9  # Adam
        self.beta2 = 0.999  # Adam
        self.epsilon = 1e-8  # Adam, RMSprop

    def predict(self, X):
        return self.W[0] * X[:, 0] + self.W[1] * X[:, 1] + self.W[2]
```

```python
# Chuyển thành DataFrame
data = pd.DataFrame({'X1': X1, 'X2': X2, 'Y': Y})
# Chia tập dữ liệu thành tập huấn luyện và kiểm tra
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
print(train_data, test_data)
```

```python
# 4. Hàm mất mát
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))
print(mse)
print(mae)
```
